```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

# Aggreagated transaction
import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
```

```python
Agg_Trans.to_csv("Aggregated transaction.csv",index = False)
```

```python
# Aggreagte user

path_user = "/content/pulse/data/aggregated/user/country/india/state/"
user_list = os.listdir(path_user)

clm_user = {'State': [], 'Year': [], 'Quarter': [], 'Brand': [], 'User_Count': [], 'Percentage': []}

for state in user_list:
    p_i = path_user + state + "/"
    for year in os.listdir(p_i):
        p_j = p_i + year + "/"
        for file in os.listdir(p_j):
            with open(p_j + file, 'r') as Data:
                D = json.load(Data)
                # Check if usersByDevice data exists (older files might vary)
                if D['data']['usersByDevice'] is not None:
                    for z in D['data']['usersByDevice']:
                        clm_user['Brand'].append(z['brand'])
                        clm_user['User_Count'].append(z['count'])
                        clm_user['Percentage'].append(z['percentage'])
                        clm_user['State'].append(state)
                        clm_user['Year'].append(year)
                        clm_user['Quarter'].append(int(file.strip('.json')))

Agg_User = pd.DataFrame(clm_user)
```

```python
Agg_User.to_csv("Aggregated user.csv",index = False)
```

```python
# Aggregated Insurance
path_ins = "/content/pulse/data/aggregated/insurance/country/india/state/"
ins_list = os.listdir(path_ins)

clm_ins = {'State': [], 'Year': [], 'Quarter': [], 'Insurance_Type': [], 'Insurance_Count': [], 'Insurance_Amount': []}

for state in ins_list:
    p_i = path_ins + state + "/"
    for year in os.listdir(p_i):
        p_j = p_i + year + "/"
        for file in os.listdir(p_j):
            with open(p_j + file, 'r') as Data:
                D = json.load(Data)
                for z in D['data']['transactionData']:
                    clm_ins['Insurance_Type'].append(z['name'])
                    clm_ins['Insurance_Count'].append(z['paymentInstruments'][0]['count'])
                    clm_ins['Insurance_Amount'].append(z['paymentInstruments'][0]['amount'])
                    clm_ins['State'].append(state)
                    clm_ins['Year'].append(year)
                    clm_ins['Quarter'].append(int(file.strip('.json')))

Agg_Ins = pd.DataFrame(clm_ins)
```

```python
# Map User

path_map_user = "/content/pulse/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path_map_user)

clm_map_user = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Registered_Users': [], 'App_Opens': []}

for state in map_user_list:
    p_i = path_map_user + state + "/"
    for year in os.listdir(p_i):
        p_j = p_i + year + "/"
        for file in os.listdir(p_j):
            with open(p_j + file, 'r') as Data:
                D = json.load(Data)
                # Map user data is stored as a dictionary under 'hoverData'
                for district, info in D['data']['hoverData'].items():
                    clm_map_user['District'].append(district)
                    clm_map_user['Registered_Users'].append(info['registeredUsers'])
                    clm_map_user['App_Opens'].append(info['appOpens'])
                    clm_map_user['State'].append(state)
                    clm_map_user['Year'].append(year)
                    clm_map_user['Quarter'].append(int(file.strip('.json')))

Map_User = pd.DataFrame(clm_map_user)
```

```python
Map_User.to_csv("Map_User.csv",index = False)
```

```python
# Map_map

path_map_tran = "/content/pulse/data/map/transaction/hover/country/india/state/"
map_tran_list = os.listdir(path_map_tran)

clm_map_tran = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Transaction_Count': [], 'Transaction_Amount': []}

for i in map_tran_list:
    p_i = path_map_tran + i + "/"
    for j in os.listdir(p_i):
        p_j = p_i + j + "/"
        for k in os.listdir(p_j):
            with open(p_j + k, 'r') as Data:
                D = json.load(Data)
                for z in D['data']['hoverDataList']:
                    clm_map_tran['District'].append(z['name'])
                    clm_map_tran['Transaction_Count'].append(z['metric'][0]['count'])
                    clm_map_tran['Transaction_Amount'].append(z['metric'][0]['amount'])
                    clm_map_tran['State'].append(i)
                    clm_map_tran['Year'].append(j)
                    clm_map_tran['Quarter'].append(int(k.strip('.json')))

Map_Trans = pd.DataFrame(clm_map_tran)
```
